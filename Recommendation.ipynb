```python
# Merging the books and ratings on ISBN
ratings_with_title = ratings.merge(books, on ='ISBN')
```

```python
#Displaying the results
ratings_with_title.head(3).style.format({'image_url': show_image})
```

```python
# Finding number of ratings of each books
num_rating_df= ratings_with_title.groupby('mod_title').count()['ratings'].reset_index()
num_rating_df.rename(columns= {'ratings':'num_ratings'},inplace=True)
num_rating_df
```

## POPULARITY BASED RECOMMENDATION

```python
avg_rating_df= ratings_with_title.groupby('mod_title').mean()['ratings'].reset_index()
avg_rating_df.rename(columns= {'ratings':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_books= num_rating_df.merge(avg_rating_df, on='mod_title')
popular_books.head()
```

```python
popular_books= popular_books[popular_books['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
popular_books.head().style.format({'image_url': show_image})
```

## COLLABORATIVE FILTERING BASED RECOMMENDATION

```python
#Considering the users who gave >200 ratings and books having >=50 ratings
req_users= ratings_with_title.groupby('user_id').count()['ratings']>200
req_users_index= req_users[req_users].index
```

```python
filtered_users= ratings_with_title[ratings_with_title['user_id'].isin(req_users_index)]
filtered_users
```

```python
req_books= filtered_users.groupby('title').count()['ratings']>=50
req_books_index= req_books[req_books].index
```

```python
final_ratings= filtered_users[filtered_users['title'].isin(req_books_index)]
final_ratings
```

```python
pt= final_ratings.pivot_table(index='title', columns='user_id',values='ratings')
pt.fillna(0,inplace=True)
pt
```

```python
similarity_score= cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  index= np.where(pt.index==book_name)[0][0]
  similar_items= sorted(enumerate(similarity_score[index]), key=lambda x:x[1], reverse=True)[1:15]

  data=[]
  for i in similar_items:
    item = []
    temp_df = books[books['title'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('title')['title'].values))
    item.extend(list(temp_df.drop_duplicates('title')['author'].values))
    item.extend(list(temp_df.drop_duplicates('title')['image_url'].values))
        
    data.append(item)
    recommendation_list= pd.DataFrame(data, columns=['title', 'author', 'image_url']).style.format({'image_url': show_image})
  return recommendation_list


#recommendation_list= pd.DataFrame(data, columns=['title', 'author', 'image_url'])
```
