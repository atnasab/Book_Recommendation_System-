```python
#Adding a modified tile column for removing necessary stop words
books['mod_title'] = books['title'].str.replace("[^a-zA-Z0-9 ]", "", regex=True)
```

```python
books
```

```python
# Lower casing the modified title column 
books['mod_title'] = books['mod_title'].str.lower()
```

```python
# Striping the unnecessary extra spaces if any
books['mod_title'] = books['mod_title'].str.replace("\s+"," " , regex=True)
```

```python
books
```

```python
books = books[books['mod_title'].str.len() > 0]
```

```python
books.to_csv('books.csv')
```

```python
#Importing TF-IDF for creating a optimised search 
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

tfidf = vectorizer.fit_transform(books['mod_title'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import re
#Displaying image rather than the links to the image
def show_image(val):
    return '<img src = "{}" width =50 ></img>'.format(val)

def search(query, vectorizer):
    processed = re.sub("[^a-zA-Z0-9 ]", "", query.lower())
    query_vec = vectorizer.transform([processed])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -100)[-100:]
    results = books.iloc[indices]
    #results = results.sort_values('ratings', ascending = False)
    return results.head(5).style.format({'image_url': show_image})
```

```python
#Search on the data for results
search("The Da Vinci Code", vectorizer)
```
